```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips') # Dataset de propinas en formato pandas
tips.head()
```

```python
sns.countplot(data=tips,x='day',hue='sex')
plt.show()
```

```python
sns.countplot(data=tips,y='day',hue='sex')
plt.show()
```

```python
sns.stripplot(data=tips, x='day', y='total_bill', hue='sex') # Muy bueno. Un grafico de dispersi√≥n pero para trabajar con variables categoricas. 
plt.show()
```

```python
sns.stripplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True) # Con Dodge=True logro que los datos separados se vean en scatter individuales. 
plt.show()
```

```python
# con  plt.figure puedo setear los tama√±os de mis lienzos en seaborn porque esta ultima est√° construida sobre matplotlib
plt.figure(figsize=(6,6))
sns.stripplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True) 
plt.show()
```

```python
# El swarmplot es muy parecido al stripplot solo que cuando la frecuencia se acumula los puntos se suman a la izquierda y derecha. 
plt.figure(figsize=(6,6))
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True) 
plt.show()
```

```python
# Puedo usar la x en el boxplot para abrir el mismo por una variable categorica. 
plt.figure(figsize=(6,6))
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True) 
plt.show()
```

```python
# Puedo sumar una segunda grafica en el mismo lienzo poniendo el codigo debajo 
plt.figure(figsize=(6,6))
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True) 
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True, color='0', marker="<") 
plt.show()
```

```python
# Muy parecido al boxplot pero sin marcas de la mediana y los percentiles. Solo me muestra la concentraci√≥n de mis datos mediante el ensanchamiento
plt.figure(figsize=(6,6))
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True, palette='pastel') 
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True, split=True, palette='pastel') # Con split True unifico el aperturado del hue en un solo grafico
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.violinplot(data=tips, y='total_bill', hue='sex', dodge=True, split=True, palette='pastel') # Con split True unifico el aperturado del hue en un solo grafico
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.violinplot(data=tips, x='total_bill', hue='sex', dodge=True, split=True, palette='pastel') # Con split True unifico el aperturado del hue en un solo grafico
plt.show()
```

```python
# Pensado para trabajar con variables categoricas
sns.catplot(data=tips, x='day', y='total_bill', hue='sex') # Usando una variable categorica y una numerica nos va a arrojar un stripplot
plt.show()
```

```python
# Pensado para trabajar con variables categoricas
sns.catplot(data=tips, x='day', y='total_bill', hue='sex', kind='swarm') # definiendo un kind='swarm' entonces el catplot nos va a dar un swarmplot
plt.show()
```

```python
# Pensado para trabajar con variables categoricas
sns.catplot(data=tips, x='day', y='total_bill', hue='sex', kind='box') # definiendo un kind='box' entonces el catplot nos va a dar un boxplot
plt.show()
```

```python
# parametro 'col=variable categorica: Va a abrir el grafico en dos o en la cantidad de valores que asuma la variable categorica que usemos. 
sns.catplot(data=tips, x='day', y='total_bill', hue='sex', kind='box', col='time') # definiendo un kind='box' entonces el catplot nos va a dar un boxplot
plt.show()
```

```python
# parametro 'col=variable categorica: Va a abrir el grafico en dos o en la cantidad de valores que asuma la variable categorica que usemos. 
sns.catplot(data=tips, x='time', y='total_bill', hue='sex', kind='box', col='day') # definiendo un kind='box' entonces el catplot nos va a dar un boxplot
plt.show()
```
