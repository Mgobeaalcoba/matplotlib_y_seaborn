```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips') # Dataset de propinas en formato pandas
tips.head()
```

```python
tips.corr() # función de Pandas para ver el nivel de correlación entre las variables numericas del dataset. Estructura matricial. Va de -1 a 1 
```

```python
sns.heatmap(tips.corr())
plt.show()
```

```python
# Damos contexto al grafico
sns.heatmap(tips.corr(), annot=True) # Sumo los valores al heatmap
plt.show()
```

```python
# Damos contexto al grafico
sns.heatmap(tips.corr(), annot=True, cmap='coolwarm') # Cambio la gama de colores. Este es termico
plt.show()
```

```python
# Damos contexto al grafico
sns.heatmap(tips.corr(), annot=True, cmap='coolwarm', linewidths=5) # Separo las cajas con una distancia seteada entre cada una
plt.show()
```

```python
# Damos contexto al grafico
sns.heatmap(tips.corr(), annot=True, cmap='coolwarm', linewidths=5, linecolor='black') # Pinto de color las lineas entre las cajas
plt.show()
```

```python
# Damos contexto al grafico
sns.heatmap(tips.corr(), annot=True, cmap='coolwarm', linewidths=5, linecolor='black',
            vmin=0, vmax=1) # Seteo los valores minimos y maximos de mi escala. De esta forma pierdop los frios en este caso porque no tengo valores inferiores a 0,49
plt.show()
```

```python
# Damos contexto al grafico
sns.heatmap(tips.corr(), annot=True, cmap='coolwarm', linewidths=5, linecolor='black',
            vmin=-1, vmax=1) # Seteo los valores minimos y maximos de mi escala. De esta forma pierdop los frios en este caso porque no tengo valores inferiores a 0,49
plt.show()
```

```python
# Damos contexto al grafico
sns.heatmap(tips.corr(), annot=True, cmap='coolwarm', linewidths=5, linecolor='black',
            vmin=0.49, vmax=1) # Seteo los valores minimos y maximos de mi escala. De esta forma pierdop los frios en este caso porque no tengo valores inferiores a 0,49
plt.show()
```

```python
# Damos contexto al grafico
sns.heatmap(tips.corr(), annot=True, cmap='coolwarm', linewidths=5, linecolor='black',
            vmin=0.49, vmax=1, cbar=False) # Apago la escala de colores
plt.show()
```
