```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips') # Dataset de propinas en formato pandas
tips.head()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.show()
# Parece verse una cierta correlación positiva entre total de la cuenta y propina
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time') # Con style puedo atar el tipo de marca del scatter a una variable categorica
plt.show()
# Parece verse una cierta correlación positiva entre total de la cuenta y propina
# Parece que para la cena se gasta mucho mas y se deja mas propina
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size') # Con size puedo atar el tamaño de marca del scatter a una variable numerica
plt.show()
```

```python
# Como la leyenda queda mal ubicada y me molesta la lectura voy a moverla
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size') # Con size puedo atar el tamaño de marca del scatter a una variable numerica
plt.legend(loc='center', bbox_to_anchor=(1.15,0.47))
plt.show()
```

```python
markers = {'Lunch':'D', 'Dinner':'s'} # Defino los markers que quiero usar para cada caso con un diccionario que ate los valores de la variable seleccionada en style a un tipo de marca
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size', markers=markers) 
plt.legend(loc='center', bbox_to_anchor=(1.15,0.47))
plt.show()
```

```python
plt.figure(figsize=(8,8)) # Seteo una medida deseada para el paño
markers = {'Lunch':'D', 'Dinner':'s'} # Defino los markers que quiero usar para cada caso con un diccionario que ate los valores de la variable seleccionada en style a un tipo de marca
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size', markers=markers) 
plt.legend(loc='center', bbox_to_anchor=(1.15,0.47))
plt.show()
```

```python
sns.lineplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size') # Dificil de analizar, pero tmb nos muestra la relación entre una variable numerica y otra
plt.legend(loc='center', bbox_to_anchor=(1.15,0.47))
plt.show()
# Prefiero siempre utilizar un scatter plot pero existe esta opción en seaborn
```

```python
# Relplot es el contrario del catplot. Un grafico pensado para relaciones numericas que puede ser modificado medidante el kind=
sns.relplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size') # Dificil de analizar, pero tmb nos muestra la relación entre una variable numerica y otra
plt.legend(loc='center', bbox_to_anchor=(1.15,0.47))
plt.show()
# Por defecto será un scatter plot pero puedo setear el estilo con el parametro kind=
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size', kind='scatter') # Dificil de analizar, pero tmb nos muestra la relación entre una variable numerica y otra
plt.legend(loc='center', bbox_to_anchor=(1.15,0.47))
plt.show()
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size', kind='line') # Dificil de analizar, pero tmb nos muestra la relación entre una variable numerica y otra
plt.legend(loc='center', bbox_to_anchor=(1.15,0.47))
plt.show()
```

```python
# Al igual que con el catplot tmb puedo usar el parametro col=variable categorica para armar mas de un grafico abierto por los valores de esa variable 
sns.relplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size', kind='line', col='time') # Dificil de analizar, pero tmb nos muestra la relación entre una variable numerica y otra
plt.legend(loc='center', bbox_to_anchor=(1.15,0.47))
plt.show()
```
