```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
y
```

```python
x
```

```python
plt.plot(x,y) # Grafico mas simple de correlación o mapa de coordenadas
plt.show()
```

```python
plt.plot(x,y, 'r') # Usando la inicial del color como parametro puedo cambiar el color de mi grafica
plt.show()
```

```python
plt.plot(x,y, 'ro')
plt.show() # Usando la inicial del color mas la letra q represente un tipo de marcado puedo cambiar color y tipo de marca a la vez
```

```python
plt.plot(x,y, 'ro-')
plt.show() # Puedo pedirle tmb puntos redondos con continuidad en el marker agregando el guion
```

```python
plt.plot(x,y, 'yD:')
plt.show() # Marker de diamantes unido por linea de puntos
```

```python
plt.hist(x)
plt.show() # plt.hist me hará un histograma de una sola variable
```

```python
plt.pie(y)
plt.show() # Grafico pie tmb es de una sola variable
```

```python
plt.scatter(x,y)
plt.show() # Scater que me muestra las uniones entre estas dos variables
```

```python
plt.boxplot(x)
plt.show() # Grafico de caja de X
```
